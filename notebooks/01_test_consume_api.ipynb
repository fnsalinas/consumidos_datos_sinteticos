```python
download_path: str = "/home/ubuntu/workspace/fsalinas/consumidos_datos_sinteticos/data/downloaded_api"
errors_number: int = 0
max_errors_number: int = 10
```

```python
while True:
    try:
        noclients: int = random.randint(1, 100)
        print((f"Downloading {noclients} clients" + " "*100)[:60], end="\r", flush=True)
        request = requests.get(f"{URL}/get_{noclients}_clients")
        if "502 Bad Gateway" in request.text:
            print("Bad Gateway" + " "*100, end="\r", flush=True)
            break
        clients = json.loads(request.text)
        ts_path: str = datetime.now().strftime("%Y%m%d_%H%M%S")
        with open(f"{download_path}/clients_{ts_path}_{noclients}.json", "w") as output_file:
            json.dump(clients, output_file, indent=4)
        random_seconsdelay: int = random.randint(60, 200)
        print((f"Waiting {random_seconsdelay} seconds" + " "*100)[:60], end="\r", flush=True)
        for second in range(random_seconsdelay):
            print((f"Waiting {second} seconds of {random_seconsdelay} seconds" + " "*100)[:100], end="\r", flush=True)
            seconds = range(240)
            time.sleep(1)
    except:
        errors_number+=1
        print(f"Error number {errors_number} occured" + " "*100, end="\r", flush=True)
        if errors_number >= max_errors_number:
            print("Max errors number reached" + " "*100, end="\r", flush=True)
            break
```

```python
json_files: List[str] = glob(f"{download_path}/*.json")
json_file: Dict[str, Any] = json_files[0]
file = open(json_file, "r")
json.loads(file.read())[0]
```

```python
json_files: List[str] = glob(f"{download_path}/*.json")

list_of_dfs_client: List[pd.DataFrame] = []
list_of_dfs_user: List[pd.DataFrame] = []
list_of_dfs_product: List[pd.DataFrame] = []
list_of_dfs_category: List[pd.DataFrame] = []
list_of_dfs_order: List[pd.DataFrame] = []
list_of_dfs_invoice: List[pd.DataFrame] = []
for json_file in json_files:
    
    with open(json_file, "r") as input_file:
        clients = json.load(input_file)

    list_clients_data: List[Any] = []
    list_users_data: List[Any] = []
    list_products_data: List[Any] = []
    list_categories_data: List[Any] = []
    list_orders_data: List[Any] = []
    list_invoices_data: List[Any] = []
    for client in clients:
        list_clients_data.append(pd.DataFrame(client["client_data"], index=[0]))
        list_users_data.append(pd.DataFrame(client["user_data"], index=[0]))
        list_products_data.append(pd.DataFrame(client["product_data"], index=[0]))
        list_categories_data.append(pd.DataFrame(client["category_data"], index=[0]))
        list_orders_data.append(pd.DataFrame(client["order_data"], index=[0]))
        list_invoices_data.append(pd.DataFrame(client["invoice_data"], index=[0]))

    list_of_dfs_client.append(pd.concat(list_clients_data, ignore_index=True))
    list_of_dfs_user.append(pd.concat(list_users_data, ignore_index=True))
    list_of_dfs_product.append(pd.concat(list_products_data, ignore_index=True))
    list_of_dfs_category.append(pd.concat(list_categories_data, ignore_index=True))
    list_of_dfs_order.append(pd.concat(list_orders_data, ignore_index=True))
    list_of_dfs_invoice.append(pd.concat(list_invoices_data, ignore_index=True))

df_client = pd.concat(list_of_dfs_client, ignore_index=True)
df_user = pd.concat(list_of_dfs_user, ignore_index=True)
df_product = pd.concat(list_of_dfs_product, ignore_index=True)
df_category = pd.concat(list_of_dfs_category, ignore_index=True)
df_order = pd.concat(list_of_dfs_order, ignore_index=True)
df_invoice = pd.concat(list_of_dfs_invoice, ignore_index=True)
```
